# 式

$$ F = \frac{1}{8}\pi^2 \rho_{air} d^3 N v_{0} $$
$$ F = mR \omega^{2} $$

$$ \omega = \sqrt{ \frac{F}{mR}} $$
$$ R = \frac{mv^{2}}{F} $$

## 初期位置

$$ p_{x_{0}} = 5.485 $$
$$ p_{y_{0}} = 0 $$
$$ p_{z_{0}} = 0.2 $$

## 位置

$$ p_{x_{i}} = R \cos ( \omega t - \theta_{0}) - R \cos \theta_{0} + p_{x_{0}} $$
$$ p_{y_{i}} = R \sin ( \omega t - \theta_{0}) - R \sin \theta_{0} + p_{y_{0}}  $$
$$ p_{z_{i}} =  -\frac{1}{2} g t^{2} + (v_{0} \sin \phi_{0}) t + p_{z_{0}} $$

## 速度

$$ v_{x} = - R \omega \sin \theta $$
$$ v_{y} =   R \omega \cos \theta $$
$$ v_{z} =  -g t + v_{0} \sin \phi_{0} $$

## 加速度

$$ a_{x} = -\frac{1}{8}\pi^2 \frac{\rho}{m} d^3 N v_{0} \cos \theta $$
$$ a_{y} = \frac{1}{8}\pi^2 \frac{\rho}{m} d^3 N v_{0} \sin \theta $$
$$ a_{z} =  -g $$

## 角度

$$ \theta\ = arctan \frac{v_{y}}{v_{x}} $$
